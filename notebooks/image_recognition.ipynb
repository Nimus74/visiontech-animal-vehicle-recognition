## Sezione 8: Inference su Nuove Immagini

Ora possiamo usare il modello addestrato per classificare nuove immagini!

### Come funziona:
1. Carica un'immagine
2. La preprocessa (resize, normalizzazione)
3. Passa attraverso il modello
4. Ottiene la predizione (Animal o Vehicle)

```python
# Funzione helper per classificare un'immagine
import torch
from PIL import Image as PILImage  # Usa alias per evitare conflitti
from torchvision import transforms
from models.cnn import SimpleCNN
from src.utils import load_config, pick_device
import io

def classify_image(image_input, model_path='outputs/checkpoints/best.pt'):
    """
    Classifica un'immagine come Animal o Vehicle.
    
    Args:
        image_input: Path dell'immagine (string) o file object (BytesIO da Colab upload)
        model_path: Path del modello addestrato
    """
    # Carica configurazione
    config = load_config('configs/default.yaml')
    device = pick_device(config['train']['device'])
    
    # Carica il modello
    model = SimpleCNN(dropout=float(config['model']['dropout'])).to(device)
    checkpoint = torch.load(model_path, map_location=device)
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()
    
    # Preprocessa l'immagine (stesso preprocessing usato nel training)
    mean = (0.4914, 0.4822, 0.4465)
    std = (0.2023, 0.1994, 0.2010)
    transform = transforms.Compose([
        transforms.Resize((32, 32)),
        transforms.ToTensor(),
        transforms.Normalize(mean, std),
    ])
    
    # Carica e trasforma l'immagine
    # Gestisce sia path string che file objects (da Colab files.upload())
    if isinstance(image_input, str):
        # Se è un path string
        img = PILImage.open(image_input).convert('RGB')
    else:
        # Se è un file object (BytesIO da Colab)
        # Se ha un metodo read(), usalo; altrimenti è già BytesIO
        if hasattr(image_input, 'read'):
            img = PILImage.open(io.BytesIO(image_input.read())).convert('RGB')
        else:
            # Se è già un BytesIO o simile
            img = PILImage.open(io.BytesIO(image_input)).convert('RGB')
    
    img_tensor = transform(img).unsqueeze(0).to(device)
    
    # Predizione
    with torch.no_grad():
        logits = model(img_tensor)
        probs = torch.softmax(logits, dim=1)
        pred = torch.argmax(logits, dim=1).item()
        confidence = probs[0][pred].item()
    
    # Risultato
    label = "Animal" if pred == 0 else "Vehicle"
    
    return label, confidence, img

print("✅ Funzione di classificazione caricata!")
```

```python
# Esempio: Classifica un'immagine dal test set
import random
from src.dataset import CIFAR10AnimalVehicle
from torchvision import transforms
from IPython.display import display

# Carica il test set
test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010))
])
test_dataset = CIFAR10AnimalVehicle(root='data', train=False, transform=test_transform, download=False)

# Prendi un'immagine casuale
idx = random.randint(0, len(test_dataset) - 1)
img, true_label, cifar_label = test_dataset[idx]

# Salva temporaneamente per la classificazione
temp_path = '/tmp/test_image.png'
from torchvision.utils import save_image
save_image(img, temp_path)

# Classifica
pred_label, confidence, _ = classify_image(temp_path)

# Visualizza risultato
print(f"🖼️  Immagine #{idx} dal test set")
print(f"📋 Label vera: {'Animal' if true_label == 0 else 'Vehicle'}")
print(f"🤖 Predizione: {pred_label}")
print(f"📊 Confidence: {confidence*100:.2f}%")

# Mostra l'immagine originale (denormalizzata)
denormalize = transforms.Normalize(
    mean=[-0.4914/0.2023, -0.4822/0.1994, -0.4465/0.2010],
    std=[1/0.2023, 1/0.1994, 1/0.2010]
)
img_display = denormalize(img)
img_display = torch.clamp(img_display, 0, 1)
display(transforms.ToPILImage()(img_display))
```
